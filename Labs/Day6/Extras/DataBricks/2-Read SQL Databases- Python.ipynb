# Read SQL databases using JDBC

- Databricks Runtime contains JDBC drivers for Microsoft SQL Server and Azure SQL Database

## Establish connectivity to SQL Server

### Create the JDBC URL

```python
jdbcHostname = "dbsqlsvrmay22.database.windows.net"
jdbcDatabase = "dbsql"
jdbcPort = 1433
jdbcUsername = "atingupta2005"
jdbcPassword = "Azure@123456"
```

```python
jdbcUrl = "jdbc:sqlserver://{0}:{1};database={2}".format(jdbcHostname, jdbcPort, jdbcDatabase)
connectionProperties = {
  "user" : jdbcUsername,
  "password" : jdbcPassword,
  "driver" : "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}
print(jdbcUrl)
print(connectionProperties)
```

## Read data from JDBC

```python
pushdown_query = "(select * from employees where employee_id < 105) emp_alias"
employees_table = spark.read.jdbc(url=jdbcUrl, table=pushdown_query, properties=connectionProperties)
display(employees_table)
```
